```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

ds.replace("-",0, inplace=True)
```

```python
ds= ds[1:]
ds.columns
```

```python
ds.dtypes
```

```python
ds['Since/LastApp']= ds['Since/LastApp'].str[:4]
```

```python
ds['Debut']= ds['Debut'].str[:4]
```

```python
ds[['Seasons', 'Points', 'GamesPlayed', 'GamesWon', 'GamesDrawn',
       'GamesLost', 'GoalsFor', 'GoalsAgainst', 'Champion', 'Runner-up',
       'Third', 'Fourth', 'Fifth', 'Sixth', 'T', 'Debut', 'Since/LastApp',
       'BestPosition']] = ds[['Seasons', 'Points', 'GamesPlayed', 'GamesWon', 'GamesDrawn',
       'GamesLost', 'GoalsFor', 'GoalsAgainst', 'Champion', 'Runner-up',
       'Third', 'Fourth', 'Fifth', 'Sixth', 'T', 'Debut', 'Since/LastApp',
       'BestPosition']].apply(pd.to_numeric)
```

```python
ds.dtypes
```

```python
#Print all the teams which have started playing between 1930-1980.


ds[ds['Debut'].between(1930,1980)]
```

```python
#Print the list of teams which came Top 5 in terms of points 

ds.nlargest(5,'Points')
```

```python
def Goal_diff_count(Goals_For,Goals_Against):
    return Goals_For-Goals_Against
```

```python
ds['Goal_diff']=ds.apply(lambda x: Goal_diff_count(x['GoalsFor'],x['GoalsAgainst']),axis=1)
```

```python
ds.tail()
```

```python
ds[ds['Goal_diff']==ds['Goal_diff'].max()][['Team','Goal_diff']]
```

```python
ds[ds['Goal_diff']==ds['Goal_diff'].min()][['Team','Goal_diff']]
```

```python
# Create a new column with name “Winning Percent” and append it to the data set
ds['WinningPercent']= ds['GamesWon']/ds['GamesPlayed']*100
ds['WinningPercent']= ds['WinningPercent'].fillna(value=0)
```

```python
#Print the top 5 teams which has the highest Winning percentage
ds.nlargest(5,'WinningPercent')
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions
df= ds[['Team', 'Points','BestPosition']].groupby(by='BestPosition')
```
